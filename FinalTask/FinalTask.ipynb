```python
import pandas as pd


train_X = pd.read_csv("X_train.csv")
test_X = pd.read_csv("X_test.csv")
train_y = pd.read_csv("y_train.csv")

train_X.isnull().sum()
```

```python
train_X['salary_gross'].value_counts()
```

```python
train_X['salary_from'] = train_X['salary_from'].fillna(0)
train_X['description'] = train_X['description'].fillna('')
train_X['salary_gross'] = train_X['salary_gross'].fillna(False)
```

```python
import spacy


nlp = spacy.load("ru_core_news_lg")
```

```python
df = train_y.merge(train_X)
```

```python
import utils


test_X['description'] = utils.remove_html_tags(test_X['description'])
df['description'] = utils.remove_html_tags(df['description'])
```

```python
num_vectorized = utils.vectorize(df, nlp)

df.info()
```

```python
num_test = utils.vectorize(test_X, nlp)

test_X.info()
```

```python
from catboost import CatBoostRegressor, Pool


X = ['vector_dim_' + str(i) for i in range(num_vectorized)]
X.append('salary_from')
X.append('name')
X.append('area_id')
cat_features = ['name']
y = ['salary_to']
```

```python
X
```

```python
model = CatBoostRegressor(
	cat_features=cat_features,
	eval_metric='SMAPE',
	learning_rate=0.03,
	random_seed=42,
	verbose=100,
	iterations=22000
)

from sklearn.model_selection import train_test_split


train, test = train_test_split(df, test_size=0.2, random_state=42)

train_pool = Pool(df[X], df[y], cat_features=cat_features)
dev_pool = Pool(test[X], test[y], cat_features=cat_features)

model.fit(train_pool, eval_set=dev_pool)
```

```python
predict = model.predict(test_X[X])
utils.submit_to_csv(predict, test_X, 'submission.csv')
```
